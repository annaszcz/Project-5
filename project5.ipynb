```python
#Creating an empty sqlite database to populate with the newly created CSVs
connector = sqlite3.connect("superstore_database.db")
```

```python
# Establishing a cursor object (a tool that executes SQL code against the database)
c = connector.cursor()
```

```python
# Drop the existing sales table if it exists
c.execute('DROP TABLE IF EXISTS sales')
c.execute('DROP TABLE IF EXISTS customers')
c.execute('DROP TABLE IF EXISTS products')
c.execute('DROP TABLE IF EXISTS phipment')
```

```python

# Sales info; Primary key
c.execute('''
CREATE TABLE IF NOT EXISTS sales (
    SalesID VARCHAR(15),
    CustomerID CHAR(8),
	ProductID VARCHAR(15),   
    Sales FLOAT(4),
	Quantity INT,
	Discount FLOAT(4),
    Profit FLOAT(4),
	PRIMARY KEY (SalesID)
	FOREIGN KEY (CustomerID) REFERENCES Customer(CustomerID)
    FOREIGN KEY (ProductID) REFERENCES Products(ProductID)
);
''')

#Customers
c.execute('''
CREATE TABLE IF NOT EXISTS customers (
	CustomerID	CHAR(8),
	CustomerName	VARCHAR(30),
	Segment	VARCHAR(15),
	City	VARCHAR(15),
	State	VARCHAR(15),
    PostalCode Char(8),
    Region VARCHAR(15),
	PRIMARY KEY(CustomerID)
);
''')

# Product info
c.execute('''
CREATE TABLE IF NOT EXISTS products (
	ProductID	VARCHAR(15),
    ProductName VARCHAR(30),
	OrderID	VARCHAR(30),
    Category VARCHAR(30),
    SubCategory VARCHAR(30),
	PRIMARY KEY(ProductID)
    FOREIGN KEY (OrderID) REFERENCES Shipment(OrderID)
);
''')

# Shipment info
c.execute('''
CREATE TABLE IF NOT EXISTS orders (
	OrderID VARCHAR(30),
    OrderDate DATE,
	ShipDate DATE,
	ShipMode	VARCHAR(15),
	PRIMARY KEY(OrderID)
);
''')
```

```python
# Append data to the existing sales table
sales.to_sql('sales', connector, if_exists='replace', index=False)
customers.to_sql('customers', connector, if_exists='replace', index=False)
products.to_sql('products', connector, if_exists='replace', index=False)
shipment.to_sql('shipment', connector, if_exists='replace', index=False)
```

```python
connector = sqlite3.connect("superstore_database.db")
c = connector.cursor()
# Export all tables into csv
import csv
with open("sales.csv", mode = "w", newline = '') as f:
    writer = csv.writer(f)

    c.execute("Select * from sales")
    for row in c:
        print(row)
        writer.writerow(row)
    connector.close()
print('csv created')
```
