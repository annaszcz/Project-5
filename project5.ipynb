```python
#Creating an empty sqlite database to populate with the newly created CSVs
connector = sqlite3.connect("superstore_database.db")
```

```python
# Establishing a cursor object (a tool that executes SQL code against the database)
c = connector.cursor()
```

```python

# Sales info; Primary key
c.execute('''
CREATE TABLE IF NOT EXISTS sales (
    SalesID VARCHAR(15),
    CustomerID CHAR(8),
	ProductID VARCHAR(15),   
    Sales FLOAT(4),
	Quantity INT,
	Discount FLOAT(4),
    Profit FLOAT(4),
	PRIMARY KEY (SalesID)
	FOREIGN KEY (CustomerID) REFERENCES Customer(CustomerID)
    FOREIGN KEY (ProductID) REFERENCES Products(ProductID)
);
''')

#Customers
c.execute('''
CREATE TABLE IF NOT EXISTS customers (
	CustomerID	CHAR(8),
	CustomerName	VARCHAR(30),
	Segment	VARCHAR(15),
	City	VARCHAR(15),
	State	VARCHAR(15),
    PostalCode Char(8),
    Region VARCHAR(15),
	PRIMARY KEY(CustomerID)
);
''')

# Product info
c.execute('''
CREATE TABLE IF NOT EXISTS products (
	ProductID	VARCHAR(15),
    ProductName VARCHAR(30),
	OrderID	VARCHAR(30),
    Category VARCHAR(30),
    SubCategory VARCHAR(30),
	PRIMARY KEY(ProductID)
    FOREIGN KEY (OrderID) REFERENCES Shipment(OrderID)
);
''')

# Shipment info
c.execute('''
CREATE TABLE IF NOT EXISTS orders (
	OrderID VARCHAR(30),
    OrderDate DATE,
	ShipDate DATE,
	ShipMode	VARCHAR(15),
	PRIMARY KEY(OrderID)
);
''')
```

```python
# What are the 3 products in each product segment with the highest sales?
query = """
WITH RankedProducts AS (
    SELECT 
        p.ProductID,
        c.Segment,
        SUM(s.Sales) AS TotalSales,
        RANK() OVER (PARTITION BY c.Segment ORDER BY SUM(s.Sales) DESC) AS rnk
    FROM 
        sales s 
    JOIN 
        customers c ON c.CustomerID = s.CustomerID 
    JOIN 
        Products p ON p.ProductID = s.ProductID 
    GROUP BY 
        p.ProductID, c.Segment
)

SELECT 
    ProductID,
    Segment,
    TotalSales
FROM 
    RankedProducts
WHERE 
    rnk <= 3;
"""

# Execute the query and store the result
top_products_by_segment = pd.read_sql(query, connector)

# Display the result
print(top_products_by_segment)

```

```python
# What are the 3 best-seller products in each product segment? (Quantity-wise)

pd.read_sql("""
WITH RankedProducts AS (
    SELECT 
        p.ProductID,
        c.Segment,
        SUM(s.Quantity) AS TotalQty,
        RANK() OVER (PARTITION BY c.Segment ORDER BY SUM(s.Quantity) DESC) AS rnk
    FROM 
        sales s 
    JOIN 
        customers c ON c.CustomerID = s.CustomerID 
    JOIN 
        Products p ON p.ProductID = s.ProductID 
    GROUP BY 
        p.ProductID, c.Segment
)

SELECT 
    ProductID,
    Segment,
    TotalQty
FROM 
    RankedProducts
WHERE 
    rnk <= 3;
""", connector)
```

```python
# What are the top 3 worst-selling products in every category? (Quantity-wise)
pd.read_sql("""
WITH RankedProducts AS (
    SELECT 
        p.ProductID,
        c.Segment,
        SUM(s.Quantity) AS TotalQty,
        RANK() OVER (PARTITION BY c.Segment ORDER BY SUM(s.Quantity) ASC) AS rnk
    FROM 
        sales s 
    JOIN 
        customers c ON c.CustomerID = s.CustomerID 
    JOIN 
        Products p ON p.ProductID = s.ProductID 
    GROUP BY 
        p.ProductID, c.Segment
)

SELECT 
    ProductID,
    Segment,
    TotalQty
FROM 
    RankedProducts
WHERE 
    rnk <= 3;
""", connector)
```
